```
#Third step
#Developing of the algorithms

def preprocessingandseparation(thetextlist): #takes a list of strings as an input
  xd = []
  for thetext in thetextlist:
    lemma = thetext.split("@")[1] #takes the lemma as it's the first element in "bold"
    result = dict()
    atlemma = "@" + lemma + "@"
    differentcontextsnoline = thetext.replace("\n", " ")
    differentcontexts = differentcontextsnoline.split(atlemma)[1].split(" • ") #every context is split by the character •
    firstcontext = differentcontexts[0]
    differentcontexts.remove(firstcontext)
    firstcontextres = dict()
    unqueried = firstcontext.split(";") #every different symbolic meaning is divided by the ";" character
    queried = processrelationships(unqueried)
    firstcontextres["generalOrUnknownSymbolicMeaning"] = queried #the first context is always general or unknown symbolic meaning
    for context in differentcontexts:
      if "@" in context:
        term = context.split("@")[1] #this is a new context
        #print(term)
        atterm = "@" + term + "@"
        newcontext = context.replace(atterm, "").split(";")
        #print(newcontext)
        newqueried = processrelationships(newcontext)
        firstcontextres[term] = newqueried 

    result[lemma] = firstcontextres
    #print(result)
    xd.append(result)
  return xd #returns a list of dictionaries divided by context and symbolic meaning for each string in the input list

def processrelationships(lista): #takes as an input a list of symbolic meanings (split before by the other algorithm)
  result = []
  for el in lista:
    switch = False
    #print(el)
    for tiporelsimb in tipirelazionisimboliche: #this is the list that contains symbolic relationships
      if tiporelsimb in el: #if a symbolic relationship is found in the symbolic meaning
        #print(el)
        switch = True
        attridict = dict()
        #newel = el.split(tiporelsimb)[1]
        #print(newel)
        definitions = []
        sdefinitions = el.split(tiporelsimb)[1].split(",") #when a symbolic relationship is present, all the elements it refers to are divided by ","
        for word in sdefinitions:
          definitions.append(word.lstrip())
        #print(definitions)
        if tiporelsimb == ":": #every time there is a ":" it means that there is a variation of the original lemma i.e sword -> flaming sword:
          attridict["Variation"] = {el.split(tiporelsimb)[0].lstrip():processrelationships(definitions)} #recursively looks for other symbolic relationships
          result.append(attridict)
          break
        elif tiporelsimb == "the personification of" or tiporelsimb == "the personifications of" or tiporelsimb == "personification of": #three different ways to write personification of
          attridict["personificationOf"] = processrelationships(definitions) #recursively looks for other symbolic relationships
          result.append(attridict)
          break
        else:
          split = tiporelsimb.split(" ")
          #print(split)
          propertysplit = []
          x = 1
          propertysplit.append(split[0])
          while x < len(split):
            propertysplit.append(split[x].capitalize())
            x +=1
          propertyname = "".join(propertysplit)
          #print(propertyname)
          attridict[propertyname] = processrelationships(definitions)
          result.append(attridict)
          break
    if switch is False: #if there is no s relationship in the element it adds it to the result
      result.append(el.lstrip())
  return result
```
